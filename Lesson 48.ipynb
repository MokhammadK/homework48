```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
```

```python
df = pd.read_csv('C:\\Users\\Bruger\\Downloads\\Data_to_Transform.csv')
df
```

```python
df.describe()
```

```python
sns.histplot(df['Moderate Positive Skew'])
fig = plt.figure()
```

```python
df_log = np.log(df['Moderate Positive Skew'])
sns.histplot(df_log)
fig = plt.figure()
```

```python
df_sqrt = np.sqrt(df['Moderate Positive Skew'])
sns.histplot(df_sqrt)
fig = plt.figure()
```

```python
df_cbrt = np.cbrt(df['Moderate Positive Skew'])
sns.histplot(df_cbrt)
fig = plt.figure()
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
data = df['Moderate Positive Skew'].values.reshape(-1, 1)
```

```python
pt = PowerTransformer(method='box-cox')
df_boxcox = pt.fit_transform(data)
```

```python
sns.histplot(df_boxcox)
fig = plt.figure()
```

```python
fig = sm.qqplot(df['Moderate Positive Skew'], line='45')
plt.show()
```

```python
fig = sm.qqplot(df_log, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_sqrt, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_cbrt, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_boxcox, line='45')
plt.show()
```
